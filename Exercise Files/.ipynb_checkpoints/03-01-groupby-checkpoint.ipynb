## <center> Transform your data with `groupby`

```python
import pandas as pd
```

```python
iris = pd.read_csv('./iris.csv')
iris.head(5)
```

A simple groupby on one dimension with one aggregation for all variables

```python
# can flatten hierarchical index with reset_index()
iris.groupby(['species']).max()
# this lets the category of species have a max for each type of species
```

Multiple aggregation methods to different variables

```python
df = iris.groupby(['species']).agg({'sepal_length':['mean','min','max'],'sepal_width':'count'})
df
# the aggregate function allows for the use of multiple functions 
# with just their string name?
```

```python
df['sepal_length']
# this maintains the hierarchical indices in the column
```

Flattening hierarchical indexes

```python
df.columns = ['_'.join(col).strip() for col in df.columns.values]
df.reset_index()
df
# df.columns is a tuple of the highest tier name, followed the 
# combinations of the children classifications
```

Specify groupings prior to any aggregation

```python
groupings = iris.groupby(['species'])
groupings
```

```python
groupings.get_group('setosa').head()
# get_group() is like a WHERE clause in SQL. 
```

```python
groupings.max()
# can use aggregation functions on the pandas group object
```

```python
groupings.apply(lambda x: x.max())
```

```python
groupings.filter(lambda x: x['petal_length'].max() <5)
# filter() is like having clause in SQL
```
